```python
import numpy as np
import matplotlib.pyplot as plt 

import pandas as pd  
import seaborn as sns 

%matplotlib inline
```

```python
from sklearn.datasets import load_boston
```

```python
boston_dataset = load_boston()
```

```python
boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
```

```python
boston['MEDV'] = boston_dataset.target
```

```python
print(boston.shape)
boston.head()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(boston, test_size = 0.6, random_state=5)
```

```python
print(train.shape)
train.head()
```

```python
X_train = train.loc[:, train.columns != 'MEDV']
Y_train = train.loc[:, train.columns == 'MEDV']


print(X_train.shape)
X_train.head()
```

```python
print(Y_train.shape)
Y_train.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

```python
X_test = test.loc[:, test.columns != 'MEDV']
Y_test = test.loc[:, test.columns == 'MEDV']

print(X_test.shape)
X_test.head()
```

```python
Y_test_result = lin_model.predict(X_test)

Y_test_result
```

```python
Y_test["MEDV_PREDICTED"] = Y_test_result
```

```python
Y_test.head()
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error as MAE
from math import sqrt


mae_result = MAE(Y_test["MEDV"], Y_test["MEDV_PREDICTED"])
rms_result = sqrt(mae_result)
r2_score_result = r2_score(Y_test["MEDV"], Y_test["MEDV_PREDICTED"])

print("mae_result : ", mae_result)
print("rms_result : ", rms_result)
print("r2_score_result : ", r2_score_result)
```
